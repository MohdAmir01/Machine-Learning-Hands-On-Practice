```python
#Converting datatype of 'TotalCharges' to numeric
df['TotalCharges'] = pd.to_numeric(df.TotalCharges, errors='coerce')
```

#### DATA MANIPULATION

Encoding all Categorical Variables

```python
#replace values in columns with binary features

binary_cols=['Partner','Dependents','PhoneService','PaperlessBilling']

for i in binary_cols:
    df[i]=df[i].replace({"Yes":1,"No":0})

#Encoding column 'gender'
df['gender']=df['gender'].replace({"Male":1,"Female":0})
```

Getting Dummy Variables for other Categorical Features

```python
category_cols=['PaymentMethod','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract']

for cc in category_cols:
    dummies = pd.get_dummies(df[cc], drop_first=False)
    dummies = dummies.add_prefix("{}#".format(cc))
    df.drop(cc, axis=1, inplace=True)
    df = df.join(dummies)
```

```python
#Drop 'CustomerID' column
df.drop('customerID', axis=1, inplace=True)
```

### LOGISTIC REGRESSION MODEL

```python
x=df.drop('Churn',axis=1)
x
```

```python
y=df['Churn']
y.head()
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_model = log_reg.fit(x_train, y_train)
```

```python
log_model.coef_
```

```python
log_model.intercept_
```

#  Predictions based on Train Data

```python
#Predictions Based on Train Data
train_Predictions = log_model.predict(x_train)
train_Predictions
```

```python
#Confusion Matrix based on Train data
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_train, train_Predictions)
print(confusion_matrix)
```

```python
#Classification Report Based on Train Data
from sklearn.metrics import classification_report
print(classification_report(y_train, train_Predictions))
```

```python
#Accuracy of Logistic Model Based on Train Data
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train, train_Predictions))
```

# Predictions Based on Test Data

```python
#Predictions Based on Test Data
test_Predictions = log_model.predict(x_test)
test_Predictions
```

```python
#Confusion Matrix based on Test Data
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, test_Predictions)
print(confusion_matrix)
```

```python
#Classification Report Based on Test Data
print(classification_report(y_test, test_Predictions))
```

```python
#Accuracy of Logistic Model Based on Test Data
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, test_Predictions))
```
